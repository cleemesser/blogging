```python
import cairo
from contextlib import contextmanager
# source Preshing on programming 
# http://preshing.com/20110920/the-python-with-statement-by-example

# surface dimensions
#WIDTH=280
#HEIGHT=204
WIDTH=512
HEIGHT=512

# ang = 0.75 # original start angle
ang = 0.75 # start angle
#STOP_ANGLE = 0.12 # original looks nice
STOP_ANGLE = 0.4

@contextmanager
def saved(cr):
    cr.save()
    try:
        yield cr
    finally:
        cr.restore()

def Tree(angle):
    cr.move_to(0, 0)
    cr.translate(0, -65)
    cr.line_to(0, 0)
    cr.stroke()
    cr.scale(0.72, 0.72)
    if angle > STOP_ANGLE:
        for a in [-angle, angle]:
            with saved(cr):
                cr.rotate(a)
                Tree(angle * 0.75)


surf = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
cr = cairo.Context(surf)
cr.translate(WIDTH/2, HEIGHT-1)
cr.set_line_width(5)
cr.scale(1.5,1.5)
Tree(ang)
surf.write_to_png('tmp.png')
```

```python
import cairo
from contextlib import contextmanager
import random

# surface dimensions
#WIDTH=280
#HEIGHT=204
WIDTH=512
HEIGHT=350

# ang = 0.75 # original start angle
ang = 0.75 # start angle
#STOP_ANGLE = 0.12 # original looks nice
STOP_ANGLE = 0.15

@contextmanager
def saved(cr):
    cr.save()
    try:
        yield cr
    finally:
        cr.restore()

def Tree(angle):
    r = random.uniform(0.7,1.3) # (random.uniform()+0.5)*.2
    s = random.uniform(0.7,1.3) # (random.uniform()+0.5)*.2
    angle = angle*r
    cr.move_to(0, 0)
    cr.translate(0, -65)
    if angle <= STOP_ANGLE:
        cr.set_source_rgb(r/1.3,(r+s)/2.6,s/1.3)
    cr.line_to(0, 0)
    cr.stroke()
    cr.scale(0.72, 0.72)
    if angle > STOP_ANGLE:
        for a in [-angle,-angle/(1.0+r),angle/(1+s), angle]:
            with saved(cr):
                cr.rotate(a)
                Tree(angle * 0.75)


surf = cairo.ImageSurface(cairo.FORMAT_ARGB32, WIDTH, HEIGHT)
cr = cairo.Context(surf)
cr.translate(WIDTH/2, HEIGHT-1)
cr.set_line_width(5)
cr.scale(1.5,1.5)
Tree(ang)
surf.write_to_png('tmp-rnd.png')
Image(filename='tmp-rnd.png')
```
